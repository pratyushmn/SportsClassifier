## Downloading Example Images
Download all the images from the file examples.csv, and then verify that they can be opened

```
path = Path('/content/data/sports')
origin = Path('/content')
folder = 'examples'
file = folder + '.csv'
dest = path/folder
dest.mkdir(parents=True, exist_ok=True)
download_images(origin/file, dest, max_pics=200)
```

```
verify_images(path/folder, delete=True, max_size=500)
```

## Loading the Model

I loaded the "resnet50main.pkl" file to get everything set up for predictions.


```
learn = load_learner(origin, 'resnet50main.pkl')
```

```
learn.data.classes
```

## Making Predictions
Finally, I used the model to make predictions on new images, and it classified them all correctly. I structured the code so that the image as well as the predicted category could both be seen in the Jupyter notebook.
